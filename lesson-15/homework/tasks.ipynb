### Task 1

```python
import matplotlib.pyplot as plt
import numpy as np

# Define x values from -10 to 10
x = np.linspace(-10, 10, 400)

# Define the function f(x)
y = x**2 - 4*x + 4

# Plotting the function
plt.plot(x, y, label='f(x) = x² - 4x + 4', color='blue')

# Adding labels and title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = x² - 4x + 4')

# Show grid and legend
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
```

### task 2

```python
# Define x values from 0 to 2π
x = np.linspace(0, 2 * np.pi, 400)

# Calculate y values
sin_y = np.sin(x)
cos_y = np.cos(x)

# Create the plot
plt.plot(x, sin_y, label='sin(x)', color='red', linestyle='--', marker='o', markevery=40)
plt.plot(x, cos_y, label='cos(x)', color='green', linestyle='-', marker='s', markevery=40)

# Add title and labels
plt.title('Sine and Cosine Functions')
plt.xlabel('x (radians)')
plt.ylabel('Function value')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

### task 3

```python
# Define x values for each function
x1 = np.linspace(-10, 10, 400)            # For x^3
x2 = np.linspace(-2 * np.pi, 2 * np.pi, 400)  # For sin(x)
x3 = np.linspace(-2, 2, 400)              # For e^x
x4 = np.linspace(0, 10, 400)              # For log(x + 1)

# Set up the 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top-left: f(x) = x^3
axs[0, 0].plot(x1, x1**3, color='blue')
axs[0, 0].set_title('f(x) = x³')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')
axs[0, 0].grid(True)

# Top-right: f(x) = sin(x)
axs[0, 1].plot(x2, np.sin(x2), color='red')
axs[0, 1].set_title('f(x) = sin(x)')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')
axs[0, 1].grid(True)

# Bottom-left: f(x) = e^x
axs[1, 0].plot(x3, np.exp(x3), color='green')
axs[1, 0].set_title('f(x) = eˣ')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')
axs[1, 0].grid(True)

# Bottom-right: f(x) = log(x + 1)
axs[1, 1].plot(x4, np.log(x4 + 1), color='purple')
axs[1, 1].set_title('f(x) = log(x + 1)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```

### task 4


```python
# Generate 100 random points (uniform distribution between 0 and 10)
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Create a scatter plot with varying colors and markers
colors = np.random.rand(100)  # Random color values
markers = ['o', 's', '^', 'D', 'P', '*']  # A list of different markers

# Plotting using different markers
plt.figure(figsize=(8, 6))
for i in range(100):
    plt.scatter(x[i], y[i], color=plt.cm.viridis(colors[i]), marker=markers[i % len(markers)])

# Add title and labels
plt.title('Scatter Plot of 100 Random Points')
plt.xlabel('X Value')
plt.ylabel('Y Value')

# Add grid
plt.grid(True)

# Show plot
plt.show()
```

### task 5

```python
# Generate 1000 random values from a normal distribution (mean=0, std=1)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create the histogram
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='skyblue', edgecolor='black', alpha=0.7)

# Add title and axis labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show grid
plt.grid(True)

# Display the plot
plt.show()
```

### task 6

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

# Compute z values
z = np.cos(x**2 + y**2)

# Create 3D surface plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add colorbar
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10)

# Set axis labels and title
ax.set_title('3D Surface Plot of f(x, y) = cos(x² + y²)')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis (f(x, y))')

# Show plot
plt.show()
```

```python
# Product names and corresponding sales data
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales = [200, 150, 250, 175, 225]
colors = ['skyblue', 'salmon', 'lightgreen', 'plum', 'orange']

# Create the bar chart
plt.figure(figsize=(8, 6))
plt.bar(products, sales, color=colors)

# Add title and axis labels
plt.title('Sales Data by Product')
plt.xlabel('Products')
plt.ylabel('Sales')

# Add grid for y-axis
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```

### task 8


```python
# Time periods
time_periods = ['T1', 'T2', 'T3', 'T4']
x = np.arange(len(time_periods))

# Sample data for each category (aligned by time period)
category_a = [20, 35, 30, 35]
category_b = [25, 32, 34, 20]
category_c = [15, 18, 20, 25]

# Plotting the stacked bars
plt.figure(figsize=(8, 6))
plt.bar(x, category_a, label='Category A', color='skyblue')
plt.bar(x, category_b, bottom=category_a, label='Category B', color='salmon')
bottom_c = np.array(category_a) + np.array(category_b)
plt.bar(x, category_c, bottom=bottom_c, label='Category C', color='lightgreen')

# Customizing the plot
plt.xticks(x, time_periods)
plt.title('Stacked Bar Chart of Category Contributions Over Time')
plt.xlabel('Time Period')
plt.ylabel('Value')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show plot
plt.show()
```
